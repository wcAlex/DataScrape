```python
from google_images_download import google_images_download as gid

# Examples: https://google-images-download.readthedocs.io/en/latest/examples.html
# Argument: https://google-images-download.readthedocs.io/en/latest/arguments.html
def download_images(keywords, limit, output_dir, driver_path):
    downloader = gid.googleimagesdownload()
    
    if limit >= 100:
        downloader.download({
            'keywords': keywords,
            "limit": limit,
            'chromedriver': driver_path,
            'output_directory': output_dir
        })
    else:
        downloader.download({
            'keywords': keywords,
            "limit": limit,
            'output_directory': output_dir
        })
```

```python
import os
import cv2 as cv
import numpy as np
from os.path import exists, isfile, join

faces_dir = './output/faces'
keyword = 'george clooney'
num_search_images = 600
images_dir = './output/images'
face_img_shape = (64, 64)

FRONT_FACE_CASCADE = cv.CascadeClassifier('scripts/haarcascade_frontalface_default.xml')
PROFILE_FACE_CASCADE = cv.CascadeClassifier('scripts/haarcascade_profileface.xml')

def save_faces(img, faces, output_dir, file_id):
    check_dir_path(output_dir)

    for i in range(len(faces)):
        x, y, w, h = faces[i]
        face_img = img[y:y+h, x:x+w]
        output_file_path = join(output_dir, '{}_{}.jpeg'.format(file_id, i))

        print(output_file_path)
        face_img = cv.resize(face_img, face_img_shape)

        cv.imwrite(output_file_path, face_img)


def detect_and_save_faces(images_dir, faces_dir):
    check_dir_path(images_dir)
    check_dir_path(faces_dir)

    file_names = [f for f in os.listdir(images_dir) if isfile(join(images_dir, f))]

    for file_name in file_names:
        file_id = file_name.split('.')[0]
        img = cv.imread(join(images_dir, file_name))
        gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

        frontal_faces = FRONT_FACE_CASCADE.detectMultiScale(gray, 1.3, 5)
        save_faces(img, frontal_faces, join(faces_dir, 'frontal'), file_id)

        profile_faces = PROFILE_FACE_CASCADE.detectMultiScale(gray, 1.3, 5)
        save_faces(img, profile_faces, join(faces_dir, 'profile'), file_id)


def check_dir_path(dir_path):
    if not exists(dir_path):
        os.makedirs(dir_path)
```

```python
detect_and_save_faces(join(images_dir, keyword), join(faces_dir, keyword))
```
