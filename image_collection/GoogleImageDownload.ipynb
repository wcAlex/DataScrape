```python
from google_images_download import google_images_download as gid

def download_images(keywords, limit, output_dir, driver_path):
    downloader = gid.googleimagesdownload()
    
    if limit >= 100:
        downloader.download({
            'keywords': keywords,
            "limit": limit,
            'chromedriver': driver_path,
            'output_directory': output_dir
        })
    else:
        downloader.download({
            'keywords': keywords,
            "limit": limit,
            'output_directory': output_dir
        })
```

```python

keyword = 'george clooney'
num_search_images = 110
images_dir = './output/images'
faces_dir = './output/faces'

download_images(keyword, num_search_images, images_dir, "/Users/chi.wang/Downloads/chromedriver")
```
